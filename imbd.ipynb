## first step: import the data

```python
from keras.datasets import imdb
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10_000)
```

## {train,test}_data: 
list of word indecies for each review.  Each index can be mapped back to a specific
word.
## {train,test}_labels:
list of 1's and 0's.  one element for each movie review.
#### 0 = negative
#### 1 = positive

## decode_review(index):
decode list of numbers back to a review (can be useful for sanity checking).

```python
def decode_review(index, data = test_data):
    word_index = imdb.get_word_index()
    reverse_word_index = { val:key for key,val in word_index.items() }
    decoded_review = ' '.join([reverse_word_index.get(i-3, '?') for i in data[index]])
    return decoded_review
```

## Data transformation
Right now we have a list of numbers, but we cannot use this data directly
### we must perform the **following transformations**
#### one hot encoding:
instead of the data for each reveiw being a variable length list of integers, we are going to use *one hot encoding*. \
That means that each list will have length 10,000 where a 1 at an index indicates the presence of that word.

```python
# apply one hot encoding to sequences
import numpy as np
def vectorize_sequences(sequences, dims=10_000):
    results = np.zeros((len(sequences), dims))
    for i, sequence in enumerate(sequences):
        results[i,sequence] = 1.0
    return results
```

```python
x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)
y_train = np.asarray(train_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')
```

## Build Network
now that our data is in vectors and our labels are scalars we can create fully connected (Dense) layers

### layers and hidden layers
for this project we chose two imediate layers with 16 hidden layers each and a third layer that will output the scalar prediction for each review

lets build our model

```python
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10_000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

### compile our model with an optimizer and a loss function

```python
from keras import optimizers
from keras import losses
model.compile(optimizer=optimizers.RMSprop(lr=0.001), loss=losses.binary_crossentropy, metrics=['accuracy'])
```

## validation
In order to monitor the accuracy of our model during training, we need to remove a small section of our training data and that small section as our validation

```python
x_val = x_train[:10_000]
partial_x_train = x_train[10_000:]
y_val = y_train[:10_000]
partial_y_train = y_train[10_000:]
```

## fit the model using training data

```python
history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs=20,
                    batch_size=512,
                    validation_data=(x_val, y_val))
```

## graph the training loss and validation loss

```python
import matplotlib.pyplot as plt

history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']

epochs = range(1,21)

plt.plot(epochs, loss_values, 'bo', label='Training loss')
plt.plot(epochs, val_loss_values, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

## plot training and validation accuracy

```python
acc_values = history_dict['accuracy']
val_acc_values = history_dict['val_accuracy']

epochs = range(1,21)

plt.plot(epochs, acc_values, 'bo', label='Training accuracy')
plt.plot(epochs, val_acc_values, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.legend()

plt.show()
```

## Update our model based on our observations
In the above graph, we can see that our validation accuracy peeks around the fourth epoch. \
this means that after the fourth epoch we are *over fitting* our model \
lets recompile our model sothat we do not overfit our model. \
\
we want to re-compile our model before calling fit again to avoid incramental training.

```python
from keras import optimizers
from keras import losses
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10_000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
model.compile(optimizer=optimizers.RMSprop(lr=0.001), loss=losses.binary_crossentropy, metrics=['accuracy'])
# now refit the model using four epochs
model.fit(x_train, y_train, epochs=4, batch_size =512)
```

## now evaluate the model

```python
results = model.evaluate(x_test,y_test)
```

## and lastly we run the test to get our results!

```python
predictions = model.predict(x_test)
```

## more sanity checking.
our predicted accuracy is 88.5%. Is this true? lets see how often our model is right...

## even more sanity checking
lets gather some summary statistics to see how accurate this thing really is

#### what percent of the time is the prediction correct?

```python
success_count=counted = 0
for pred,rate in zip(predictions, test_labels):
    if pred == .5:
        continue
    elif pred < .5:
        counted+=1
        success_count += (rate == 0)
    else:
        counted += 1
        success_count += (rate == 1)
success_count/counted
```

## check accuracy when the model is very sure


```python
success_count=counted = 0
for pred,rate in zip(predictions, test_labels):
    if pred < .1:
        counted+=1
        success_count += (rate == 0)
    elif pred > .9:
        counted += 1
        success_count += (rate == 1)
success_count/counted
```

## results
as you can see, the empirical accuracy is almost identical to the validation accuracy! \
Also, if we only include very high and very low predictions, we get better than 95% accuracy.
